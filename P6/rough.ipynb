```python
import csv
```

```python
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data
```

```python
csv_rows = process_csv("airbnb.csv")

csv_header = csv_rows[0] 

csv_data = csv_rows[1:]
```

```python
# Using the cell Function from lab6 to get the cell content

def cell(row_idx, col_name):
    
    col_idx = csv_header.index(col_name)
    val = csv_data[row_idx][col_idx]
    
    int_list = ["price", "minimum_nights", "number_of_reviews", \
                "calculated_host_listings_count", "availability_365"]
    float_list = ["latitude", "longitude", "reviews_per_month"]

    if val == "":
        return None
    elif col_name in int_list:
        return int(val)
    elif col_name in float_list:
        return float(val)
    else:
        return val
    
```

```python
# One phrase

def find_room_names(phrase):
    """
    Returns a list of all the room names that contain the substring (case insensitive match)
    passed as an argument to the paramenter `phrase`. 
    """
    # TODO: create a list
    room_list = list() 
    
    for row_idx in range(len(csv_data)):
        room_name = cell(row_idx, "name") 
        
        if room_name == None:
            continue
        
        name_list = room_name.split(" ")
        
        for i in range(len(name_list)):
        
            if phrase.lower() in name_list[i].lower(): 
                room_list.append(room_name)

    return room_list
```

```python

def median(some_list):
    """
    Returns median of a list passed as arugment
    """
    sorted_list = sorted(some_list)    
    length = len(some_list)

    if length > 0:
        if length % 2 != 0:
            mid = (length - 1) // 2
            median = int(sorted_list[mid])
            return median

        else:
            mid1 = int(sorted_list[length // 2]) 
            mid2 =  int(sorted_list[(length // 2) - 1])
            median = (mid1 + mid2) / 2
            return median
    else:
        return None
```
